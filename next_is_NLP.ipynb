```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data
```

```python
test_data
```

```python
TFid = TfidfVectorizer()
X_train_tfidf = TFid.fit_transform(train_data['Paragraph'] + ' ' + train_data['Sentence'])
y_train = train_data['Is_Next']
X_train, X_val, y_train, y_val = train_test_split(X_train_tfidf, y_train, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
val_predictions = model.predict(X_val)
val_f1_score = 100*f1_score(y_val, val_predictions, average='weighted')
print(f'Validation F1 Score: {val_f1_score}')
```

```python
X_test_tfidf = TFid.transform(test_data['Paragraph'] + ' ' + test_data['Sentence'])
test_predictions = model.predict(X_test_tfidf)


submission = pd.DataFrame({'ID': test_data['ID'], 'Is_Next': test_predictions})
submission.to_csv('submission.csv', index=False)
```
